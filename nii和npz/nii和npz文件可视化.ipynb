```python
import nibabel as nib
import matplotlib.pyplot as plt

# 设置文件路径
filepath = r'C:\Users\32050\Desktop\9036770.nii'
filepath_mask = r'C:\Users\32050\Desktop\9036770_mask.nii'

# 使用 nibabel 加载 NIfTI 文件
nii      = nib.load(filepath)
nii_mask = nib.load(filepath_mask)

# 提取图像数据
nii_data = nii.get_fdata()
nii_data_mask = nii_mask.get_fdata()

# 显示数据信息
print(nii_data.shape)
print(nii_data_mask.shape)

# 可以将数据可视化，例如显示第一个切片
plt.imshow(nii_data[:, :, 2], cmap='gray')
plt.colorbar()
plt.show()

plt.imshow(nii_data_mask[:, :, 2], cmap='gray')
plt.colorbar()
plt.show()
```
